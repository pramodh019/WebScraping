```python
from selenium import webdriver
```

```python
driver=webdriver.Chrome()
```

```python
link = 'https://www.youtube.com/@NatGeo/videos'
```

```python
driver.get(link)
```

```python
Soup = BeautifulSoup(driver.page_source,'html.parser')
```

```python
print(Soup.text.replace('\n\n',''))
```

```python
sp=Soup.find_all('ytd-rich-item-renderer')
```

```python
len(sp)
```

```python
sp[0].text.replace('\n\n','')
```

```python
all_videos = []

# Loop through each video on the page
for spp in Soup.find_all('ytd-rich-item-renderer'):
    title = spp.find('a', class_='yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media').text
    video_link = 'https://www.youtube.com/' + spp.find('a', class_='yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media').get('href')
    
    sp1 = spp.find_all('span')
    
    runtime = sp1[0].text.split() 
    views = sp1[4].text 
    uploaded = sp1[5].text 
    img_link =spp.find('img').get('src') 
    
    video_data = {
        'title': title,
        'video_link': video_link,
        'runtime': runtime,
        'views': views,
        'uploaded': uploaded,
        'img_link': img_link  
    }
    
    all_videos.append(video_data)


```

```python
all_videos
```

```python
len(all_videos)
```

```python
df = pd.DataFrame(all_videos)
```
