scrapping author details

```python
for sp in soup.find_all('div',class_='quote'):

    quote = sp.find('span',class_='text').text[1:-1]
    author = sp.find('small',class_='author').text
    author_id= sp.find('a').get('href')

    tags=[]
    
    for tag in sp.find_all('a',class_='tag'):
     tags.append(tag.text)
    tags= ','.join(tags)
    
    print(quote)
    print(author)
    print(author_id)
    print(tags)
    print('*'*150)
```

saving in a list

```python
data=[]
for sp in soup.find_all('div',class_='quote'):

    quote = sp.find('span',class_='text').text[1:-1]
    author = sp.find('small',class_='author').text
    author_id= sp.find('a').get('href')

    tags=[]
    
    for tag in sp.find_all('a',class_='tag'):
     tags.append(tag.text)
    tags= ','.join(tags)
    
    data.append([quote,author,author_id,tags])
    
```

Scrapping multiple pages

```python
pip install tqdm
```

```python
from tqdm import tqdm
data = []

for page in tqdm(range(1, 11)):
    link = 'https://quotes.toscrape.com/page/' + str(page)
    res = requests.get(link)
    soup = BeautifulSoup(res.text, 'html.parser')

    for sp in soup.find_all('div', class_='quote'):
        quote = sp.find('span', class_='text').text[1:-1]
        author = sp.find('small', class_='author').text
        author_id = sp.find('a').get('href')

        tags = []
        for tag in sp.find_all('a', class_='tag'):
            tags.append(tag.text)
        tags = ','.join(tags)

        data.append([quote, author, author_id, tags])
```

```python
data
```

```python
df=pd.DataFrame(data,columns=['Quote','Author','Author_id','Tags'])
```

```python
df
```

```python
df['Author_id_link']='https://quotes.toscrape.com'+df['Author_id']
```

```python
df['Author_id_link'].reset_index(drop=True)
```

scrappping author details from multiple pages

```python
data1=[]
for author_details in df['Author_id_link']:
    res = requests.get(author_details)
    soup = BeautifulSoup(res.text, 'html.parser')
    info = soup.find('div', class_='author-details')
    data1.append(info.text)
```

```python
data1[0]
```

```python
df1=pd.DataFrame(data1,columns=['author info'])
```

```python
df1
```

```python
combined_df = pd.concat([df, df1], axis=1)
```
